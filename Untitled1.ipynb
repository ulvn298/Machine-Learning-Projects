```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[50], [100], [150], [200], [250]])
y = np.array([500000, 1000000, 1500000, 2000000, 2500000])

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[300]])
print(f"توقع سعر البيت اللي مساحته 300 متر هو: {prediction[0]} جنيه")
```
