```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = ["I love this product", "This is a bad service", "Amazing experience", "I hate it"]
labels = ["Positive", "Negative", "Positive", "Negative"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

new_text = ["I feel happy today"]
print(f"Sentiment: {model.predict(vectorizer.transform(new_text))}")
```
